```python
 def male_female_Children(passenger):
        age,sex = passenger
        
        if age < 16:
            return 'child'
        else:
            return sex
```

```python


titanic['Person'] = titanic[['Age','Sex']].apply(male_female_Children,axis=1)
```

```python
def alone_passenger(passenger):
    sibsp,parch = passenger
    
    if sibsp == 0 and parch == 0 :
        return 'Alone'
    else:
        return 'Not Alone'
```

```python
titanic['Alone'] = titanic [['SibSp','Parch']].apply(alone_passenger, axis =1)
```

```python
titanic
```

```python
sns.catplot(x ='Alone',kind ='count', data = titanic)
```

```python
sns.catplot(x='Alone', kind ='count', data= titanic, hue='Survived')
```

```python
titanic['Survior'] = titanic.Survived.map({0:'Did not Survive',1:'Survived'})
```

```python
sns.catplot(x='Alone', kind ='count', data= titanic, hue='Survior')
```
